```python
import torch
import torch.nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 4, figsize=(24, 6))

x = np.linspace(-10, 10, 50)
x = torch.from_numpy(x)

sigmoid = torch.nn.Sigmoid()
y_sigmoid = sigmoid(x)
tanh = torch.nn.Tanh()
y_tanh = tanh(x)
relu = torch.nn.ReLU()
y_relu = relu(x)
softplus = torch.nn.Softplus()
y_softplus = softplus(x)

ax[0].plot(x, y_sigmoid, 'r')
ax[0].set_title('Sigmoid')
ax[0].grid()
ax[1].plot(x, y_tanh, 'b')
ax[1].set_title('Tanh')
ax[1].grid()
ax[2].plot(x, y_relu)
ax[2].set_title('ReLU')
ax[2].grid()
ax[3].plot(x, y_softplus)
ax[3].set_title('Softplus')
ax[3].grid()
```
